```python
from matplotlib import pyplot as plt
plt.style.use("seaborn-whitegrid")
```

```python
x = [100, 1000, 10000, 100000, 1000000]
```

```python
filename = "graph"
count = 0
def plot_me(x, y_4, y_4n, y_15, y_15n, y_26, y_26n, y_label, title = "", linest_4 = ':', linest_15 = '--', linest_26 = '-'):
    global count
    plt.figure(figsize=(15, 10))
    plt.plot(x, y_4, c = 'r', label = 'Abeceda: 4, kodirano', linestyle = linest_4, lw = 4)
    plt.plot(x, y_4n, c = 'b', label = 'Abeceda: 4, ne kodirano', lw = 4, linestyle = linest_4)
    plt.plot(x, y_15, c = 'r', label = 'Abeceda: 15, kodirano', lw = 4, linestyle = linest_15)
    plt.plot(x, y_15n, c = 'b', label = 'Abeceda: 15, ne kodirano', lw = 4, linestyle = linest_15)
    plt.plot(x, y_26, c = 'r', label = 'Abeceda: 26, kodirano', lw = 4, linestyle = linest_26)
    plt.plot(x, y_26n, c = 'b', label = 'Abeceda: 26, ne kodirano', lw = 4, linestyle = linest_26)
    plt.xlabel("Duljina sekvence")
    plt.xticks(x)
    plt.xscale('log')
    plt.ylabel(y_label)
    plt.legend()
    plt.title(title)
    plt.savefig(filename=filename + str(count) + ".png", bbox_inches = "tight")
    count += 1
    plt.show()

def plot_them(x, y_4, y_15, y_26, y_label, title = "", linest_4 = '-', linest_15 = '-', linest_26 = '-'):
    global count
    plt.figure(figsize=(15, 10))
    plt.plot(x, y_4, c = 'r', label = 'Abeceda: 4, kodirano', linestyle = linest_4, lw = 4)
    plt.plot(x, y_15, c = 'k', label = 'Abeceda: 15, kodirano', lw = 4, linestyle = linest_15)
    plt.plot(x, y_26, c = 'g', label = 'Abeceda: 26, kodirano', lw = 4, linestyle = linest_26)
    plt.xlabel("Duljina sekvence")
    plt.xticks(x)
    plt.xscale('log')
    plt.ylabel(y_label)
    plt.legend()
    plt.title(title)
    plt.savefig(filename=filename + str(count) + ".png", bbox_inches = "tight")
    count += 1
    plt.show()
```

```python
# vrijeme izgradnje stabla
y_4=[61, 118, 598, 5536, 54351]  
y_4n=[61, 116, 596, 6023, 57810]

y_15=[136, 244, 1349, 9966, 111773]
y_15n=[96, 215, 1133, 14074, 105501]


y_26=[120, 280, 1516, 12819, 133642]
y_26n=[136, 293, 1478, 12490, 132749]

x_label = "Vrijeme izgradnje stabla[$\mu$s]"
plot_me(x, y_4, y_4n, y_15, y_15n, y_26, y_26n, x_label, "Vrijeme izgradnje stabla za umjetno stvorene nizove")
```
